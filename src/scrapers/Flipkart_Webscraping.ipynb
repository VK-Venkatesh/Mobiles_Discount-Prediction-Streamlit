```python
urls = []

import time

# Handling url duplication for same product
from urllib.parse import urlparse, urlunparse
def clean_url(url):
    parsed = urlparse(url)
    return urlunparse((parsed.scheme, parsed.netloc, parsed.path, '', '', ''))

for page in range(2, 42):  # Adjust range for 500 products (20â€“25 pages)
    print(f"Scraping Page {page}...")

    product_links = driver.find_elements(By.XPATH, "//a[contains(@href, '/p/')]")
    
    before = len(urls)
    for link in product_links:
        url = link.get_attribute("href")
        url = clean_url(url)
        if url and url not in urls:
            urls.append(url)
    after = len(urls)

    print(f"Collected Ads Urls: {after - before}")

    nexturl = f"https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&q=mobiles&otracker=categorytree&page={page}"

    # # Scroll down to load Next button
    # driver.execute_script("window.scrollTo(0, document.body.scrollHeight-5000);")
    # # Click "Next"
    try:
        driver.get(nexturl)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight-5000);")
        time.sleep(2)
    except Exception as e:
        print("Could not find/click Next:", st(e))
        break
print("Total Collected URLs:", len(urls))
```
