```python
import time
import random
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Custom headers to avoid blocking
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-IN,en;q=0.9"
}

product_list = []

# Loop through multiple pages to get 500 items
for page in range(1, 26):  # Up to 25 pages
    url = f"https://www.amazon.in/s?k=mobiles&page={page}"
    print(f"🔎 Scraping page {page}...")

    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        print(f"⚠️ Failed to retrieve page {page}")
        continue

    soup = BeautifulSoup(response.content, "html.parser")

    # Select all product cards
    items = soup.select("div.s-main-slot div[data-component-type='s-search-result']")

    print(f"📦 Products found: {len(items)}")

    for item in items:
        try:
            title = item.select_one("h2 span").text.strip()
        except:
            title = ""

        try:
            mrp = item.select_one(".a-price.a-text-price span:nth-of-type(2)").text.strip()
        except:
            mrp = ""

        try:
            price = item.select_one(".a-price-whole").text.strip()
        except:
            price = ""

        try:
            discount_tag = item.find("span", string=lambda t: t and "% off" in t)
            discount = discount_tag.text.strip() if discount_tag else ""
        except:
            discount = ""

        try:
            rating = item.select_one(".a-icon-alt").text.strip()
        except:
            rating = ""

        try:
            review = item.select_one(".s-underline-text").text.strip()
        except:
            review = ""

        brand = title.split()[0] if title else ""

        product_list.append({
            "Product Name": title,
            "MRP": mrp,
            "Selling Price": price,
            "Discount %": discount,
            "Brand": brand,
            "Rating": rating,
            "Review Count": review,
            "Platform": "Amazon"
        })

    print(f"✅ Page {page} done. Total collected: {len(product_list)}")
    time.sleep(random.uniform(2, 4))  # Polite delay

    if len(product_list) >= 500:
        break

# Save to CSV
df = pd.DataFrame(product_list[:500])
df.to_csv("amazon_mobiles_500.csv", index=False, encoding='utf-8')
print("🎯 Done! Saved 500 products to amazon_mobiles_500.csv")
```
